### Question
Create a dictionary that maps names to ages. Add a new entry, remove an entry, check if a key exists, and iterate over keys, values, and items.

**Hint:** Use `.get()` for safe access and `.items()` for key-value iteration.

## Exercise 3 — List Comprehensions vs Loops

### Question
Convert the following loop into a list comprehension:
```python
squares = []
for x in range(10):
    if x % 2 == 0:
        squares.append(x ** 2)
```

**Hint:** Syntax is `[expression for item in iterable if condition]`.

<details>
<summary>Click to view solution</summary>

```python
squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(squares)  # [0, 4, 16, 36, 64]
```
</details>

## Exercise 5 — Generators

### Question
Write a generator `lazy_evens(n)` that yields even numbers from `0` to `n`. Test it by iterating and collecting results into a list.

**Hint:** Use `yield` instead of `return`. Generators are lazy and can only be iterated once.

<details>
<summary>Click to view solution</summary>

```python
def lazy_evens(n):
    i = 0
    while i <= n:
        if i % 2 == 0:
            yield i
        i += 1

# Test
print(list(lazy_evens(10)))  # [0, 2, 4, 6, 8, 10]
```
</details>

## Exercise 7 — Truthiness and Safe Access

### Question
Use Python's truthiness to safely extract the first character of a string `s`, returning `""` if `s` is `None` or empty.

**Hint:** `s and s[0]` returns `s[0]` if `s` is truthy, otherwise returns `s`. Use `or ""` to guarantee a string.

<details>
<summary>Click to view solution</summary>

```python
def safe_first_char(s):
    return (s and s[0]) or ""

print(repr(safe_first_char("Hello")))  # 'H'
print(repr(safe_first_char("")))       # ''
print(repr(safe_first_char(None)))     # ''
```
</details>

## Exercise 8 — Count Word Frequency

### Question
Count how many times each word appears in a list using both a regular dictionary and `defaultdict(int)`. Compare the code complexity.

**Hint:** Regular dict requires `if key in dict` checks. `defaultdict(int)` auto-initializes missing keys to `0`.

<details>
<summary>Click to view solution</summary>

```python
from collections import defaultdict

words = ["data", "science", "data", "python", "data"]

# Regular dict
freq1 = {}
for w in words:
    freq1[w] = freq1.get(w, 0) + 1

# defaultdict
freq2 = defaultdict(int)
for w in words:
    freq2[w] += 1

print("Regular:", dict(freq1))
print("defaultdict:", dict(freq2))
```
</details>

## Exercise 9 — `zip` and Argument Unpacking

### Question
1. Use `zip()` to pair names and scores into a list of tuples.
2. Use argument unpacking (`*`) to unzip them back into two separate lists.

**Hint:** `zip(list1, list2)` pairs items. `zip(*paired)` unzips them.

<details>
<summary>Click to view solution</summary>

```python
names = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 92]

# Pair
paired = list(zip(names, scores))
print("Paired:", paired)  # [('Alice', 90), ('Bob', 85), ('Charlie', 92)]

# Unzip
names2, scores2 = zip(*paired)
print("Unzipped names:", list(names2))
print("Unzipped scores:", list(scores2))
```
</details>

## Exercise 10 — `enumerate`

### Question
Iterate over a list of documents and print the index and document name using `enumerate()`. Then print only the index for documents containing the letter "a".

**Hint:** `enumerate(iterable, start=0)` yields `(index, value)` pairs.

## Exercise 12 — `functools.partial`

### Question
Use `partial` to create a function `times_ten` that multiplies any number by 10, and a function `power_of_2` that squares any number.

**Hint:** `partial(func, fixed_arg)` pre-fills arguments.

<details>
<summary>Click to view solution</summary>

```python
from functools import partial

def multiply(a, b):
    return a * b

def power(base, exp):
    return base ** exp

times_ten = partial(multiply, b=10)
power_of_2 = partial(power, exp=2)

print(times_ten(5))   # 50
print(power_of_2(7))  # 49
```
</details>
