## Experiment 3 — Truthiness & Short-Circuiting

```python
# Test how Python evaluates different values in boolean contexts
test_values = [0, 0.0, "", [], {}, set(), None, False, 1, "hello", [1, 2], {"a": 1}]

print("Falsy values in Python:")
for val in test_values:
    if not val:
        print(f"  {repr(val):<15} → {bool(val)}")

print("\nTruthy values in Python:")
for val in test_values:
    if val:
        print(f"  {repr(val):<15} → {bool(val)}")

# Explore short-circuit behavior of `and` / `or`
print("\nShort-circuit examples:")
print(f"0 or 'default'       → {0 or 'default'}")
print(f"'hello' and 'world'  → {'hello' and 'world'}")
print(f"[] or [1, 2, 3]      → {[] or [1, 2, 3]}")
print(f"None and print('hi') → {None and print('hi')}")  # print never runs
```

## Experiment 8 — Comprehensions vs Loops vs Functional Approaches

```python
numbers = list(range(10))

# 1. Traditional loop
squares_loop = []
for n in numbers:
    squares_loop.append(n ** 2)

# 2. List comprehension
squares_comp = [n ** 2 for n in numbers]

# 3. map + lambda
squares_map = list(map(lambda n: n ** 2, numbers))

print(f"All equal: {squares_loop == squares_comp == squares_map}")

# Comprehension with condition
even_squares = [n ** 2 for n in numbers if n % 2 == 0]
print(f"Even squares: {even_squares}")

# Nested comprehension
pairs = [(x, y) for x in range(3) for y in range(3)]
print(f"Cartesian product: {pairs}")
```

### 💡 Try modifying:

## Experiment 10 — Generators & Lazy Evaluation

```python
import sys

# Compare memory footprint: list vs generator
big_list = [x for x in range(1_000_000)]
big_gen = (x for x in range(1_000_000))

print(f"List size: {sys.getsizeof(big_list):,} bytes")
print(f"Gen size:  {sys.getsizeof(big_gen):,} bytes")

# Custom generator function
def lazy_range(n):
    i = 0
    while i < n:
        yield i
        i += 1

generator = lazy_range(5)
print(f"\nGenerator object: {generator}")

print("First pass:")
for value in generator:
    print(value, end=" ")

# ⚠️ Generators can only be consumed once!
print("\n\nSecond pass (empty):")
print(list(generator))
```

### 💡 Try modifying:

- Compare `sys.getsizeof(range(1_000_000))` with list and generator

## Experiment 11 — `*args`, `**kwargs` & Functional Flexibility

```python
# Build a flexible function wrapper that logs calls
def log_calls(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        print(f"  args:   {args}")
        print(f"  kwargs: {kwargs}")
        result = func(*args, **kwargs)
        print(f"  Result: {result}\n")
        return result
    return wrapper

@log_calls
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

@log_calls
def add_many(*numbers, multiplier=1):
    return sum(numbers) * multiplier

# Test
greet("Anuj")
greet("Anuj", greeting="Hey")
add_many(1, 2, 3, multiplier=10)
```

## Experiment 12 — `zip()`, `enumerate()` & Argument Unpacking

```python
names = ["Alice", "Bob", "Charlie"]
scores = [92, 87, 95]

# zip pairs items
print("Using zip():")
for name, score in zip(names, scores):
    print(f"  {name}: {score}")

# enumerate adds index
print("\nUsing enumerate():")
for idx, name in enumerate(names, start=1):
    print(f"  {idx}. {name}")

# Unpacking with zip
paired = list(zip(names, scores))
print(f"\nPaired: {paired}")

names2, scores2 = zip(*paired)
print(f"Unzipped names: {list(names2)}")
print(f"Unzipped scores: {list(scores2)}")
```

## Experiment 13 — Exception Handling & Defensive Programming

```python
def safe_division():
    try:
        number = int(input("Enter a number: "))
        result = 10 / number
        print(f"10 / {number} = {result}")
    except ZeroDivisionError:
        print("Cannot divide by zero!")
    except ValueError:
        print("Invalid input — please enter a number.")
    except Exception as e:
        print(f"Unexpected error: {e}")
    finally:
```

## Experiment 14 — Classes & Object-Oriented Thinking

```python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def average(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0

    def introduce(self):
        return f"{self.name} studies {self.course}"

# Create instances
student1 = Student("Anuj", "Data Science")
student1.add_score(92)
student1.add_score(88)

student2 = Student("Priya", "Machine Learning")
student2.add_score(95)

print(student1.introduce())
print(f"  Scores: {student1.scores}, Average: {student1.average()}")
print(student2.introduce())
print(f"  Scores: {student2.scores}, Average: {student2.average()}")
```
