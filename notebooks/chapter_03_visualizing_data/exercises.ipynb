## Exercise 3 — Product Sales Bar Chart

### Question

Visualize sales for different products and identify top and bottom performers.

**Data:**
```python
products = ["Laptop", "Phone", "Tablet", "Monitor"]
sales = [120, 200, 80, 150]
```

**Interpretation questions:**
- Which product sold the most?
- Which sold the least?
- How large are the differences between products?

<details>
<summary>Click to view solution</summary>

```python
products = ["Laptop", "Phone", "Tablet", "Monitor"]
sales = [120, 200, 80, 150]
colors = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']

bars = plt.bar(products, sales, color=colors, edgecolor='black', linewidth=1.2)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 2,
             str(height), ha='center', va='bottom', fontweight='bold')

plt.title("Product Sales", fontsize=14, fontweight='bold')
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.ylim(0, max(sales) * 1.15)
plt.show()
```
</details>

## Exercise 6 — Building a Histogram from Scratch (Manual Bucketing)

### Question

Given a list of student ages, bucket them into decades (20s, 30s, etc.), count frequencies, and plot as a centered bar chart. Label axes appropriately.

**Data:**
```python
ages = [22, 25, 31, 29, 38, 42, 24, 35, 30, 28, 45, 33, 27, 39, 41]
```

**Hint:** Use `Counter`, a lambda for bucketing, and offset bars by `-4` with `width=8` to center on decade boundaries.

<details>
<summary>Click to view solution</summary>

```python
from collections import Counter

ages = [22, 25, 31, 29, 38, 42, 24, 35, 30, 28, 45, 33, 27, 39, 41]

# Bucket by decade
decade = lambda age: (age // 10) * 10
age_counts = Counter(decade(age) for age in ages)

# Plot histogram with centered bars
plt.figure(figsize=(8, 5))
plt.bar([x - 4 for x in age_counts.keys()], age_counts.values(),
        width=8, edgecolor='black', color='orange')
plt.axis([15, 55, 0, max(age_counts.values()) + 1])
plt.xticks([20, 30, 40, 50])
plt.xlabel("Age Decade")
plt.ylabel("Number of Students")
plt.title("Age Distribution", fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
</details>

## Exercise 8 — Scatter Plot with Annotations & Equal Axes

### Question

Plot height vs weight for 5 individuals. Annotate each point with their initials. Then demonstrate why `plt.axis("equal")` matters by comparing side-by-side.

**Data:**
```python
heights = [160, 175, 180, 165, 190]  # cm
weights = [55, 70, 80, 60, 95]       # kg
initials = ['A', 'B', 'C', 'D', 'E']
```

**Hint:** Use `plt.subplot()` for side-by-side comparison, `plt.scatter()`, and `plt.annotate()`.

<details>
<summary>Click to view solution</summary>

```python
heights = [160, 175, 180, 165, 190]
weights = [55, 70, 80, 60, 95]
initials = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(12, 5))

# Without equal axes
plt.subplot(1, 2, 1)
plt.scatter(heights, weights, s=100, alpha=0.7, color='salmon', edgecolors='black')
for i, txt in enumerate(initials):
    plt.annotate(txt, (heights[i], weights[i]), xytext=(5, 5),
                 textcoords='offset points', fontweight='bold')
plt.title("Default Axes", fontweight='bold')
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True, alpha=0.3)

# With equal axes
plt.subplot(1, 2, 2)
plt.scatter(heights, weights, s=100, alpha=0.7, color='steelblue', edgecolors='black')
for i, txt in enumerate(initials):
    plt.annotate(txt, (heights[i], weights[i]), xytext=(5, 5),
                 textcoords='offset points', fontweight='bold')
plt.axis("equal")
plt.title("Equal Axes (True Proportions)", fontweight='bold')
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
</details>

## Exercise 9 — Fixing a Misleading Visualization

### Question

Given two years of user counts that increased only slightly, create two subplots:
1. **Misleading**: y-axis truncated to exaggerate the change
2. **Honest**: y-axis starts at 0

Add titles explaining the difference.

**Data:**
```python
years = [2022, 2023]
users = [1000, 1050]
```

**Hint:** Use `plt.axis([xmin, xmax, ymin, ymax])` to control axis ranges.

<details>
<summary>Click to view solution</summary>

```python
years = [2022, 2023]
users = [1000, 1050]

plt.figure(figsize=(12, 5))

# Misleading
plt.subplot(1, 2, 1)
plt.bar([2021.6, 2022.6], users, width=0.8, color='salmon', edgecolor='black')
plt.xticks(years)
plt.ylabel("# of Users")
plt.axis([2021.5, 2023.5, 950, 1100])  # Truncated y-axis
plt.title("Misleading: Truncated Y-Axis", fontweight='bold')
plt.grid(axis='y', alpha=0.3)

# Honest
plt.subplot(1, 2, 2)
plt.bar([2021.6, 2022.6], users, width=0.8, color='lightgreen', edgecolor='black')
plt.xticks(years)
plt.ylabel("# of Users")
plt.axis([2021.5, 2023.5, 0, 1200])  # Starts at 0
plt.title("Honest: Y-Axis Starts at 0", fontweight='bold')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
</details>

## Exercise 11 — Professional Chart Challenge

### Question

Create a polished bar chart showing project completion rates. Include:
- Custom colors per bar
- Grid lines (y-axis only)
- Value labels on top of each bar
- Bold title and axis labels
- Proper spacing to prevent cutoff

**Data:**
```python
projects = ['Alpha', 'Beta', 'Gamma', 'Delta']
completion = [85, 92, 78, 95]
```

**Hint:** Loop through `plt.bar()` return values, use `bar.get_height()` and `plt.text()`, and apply `plt.tight_layout()`.

<details>
<summary>Click to view solution</summary>

```python
projects = ['Alpha', 'Beta', 'Gamma', 'Delta']
completion = [85, 92, 78, 95]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

plt.figure(figsize=(10, 6))
bars = plt.bar(projects, completion, color=colors, edgecolor='black', linewidth=1.2)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height}%', ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.ylabel("Completion %", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
```

## Exercise 12 — Understanding Misleading Visualizations

### Question

Why can graphs sometimes be misleading? Think about:
- Axis scaling
- Missing labels
- Wrong chart choice
- Too much information
- Truncated axes

<details>
<summary>Click to view solution</summary>

Visualizations become misleading when:

1. **Axes are manipulated** — Truncating the y-axis exaggerates small differences and creates false impressions of dramatic change
2. **Labels are unclear or missing** — Viewers can't interpret what the data represents
3. **Wrong chart types are used** — Using a line chart for categorical data or a pie chart with too many slices obscures insights
4. **Too much clutter** — Excessive decoration, too many data series, or overcrowded visuals hide the key message
5. **Scales are inconsistent** — Not using `axis("equal")` for comparable variables distorts relationships
6. **Context is omitted** — Showing data without baselines, benchmarks, or historical context can lead to wrong conclusions

**Golden rule:** Good visualizations prioritize clarity and honest representation over visual flair.
</details>
