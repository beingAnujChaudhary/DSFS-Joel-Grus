## Exercise 1 — Vector Addition

### Question

Write a function that adds two vectors by combining corresponding elements.

**Example:**
```python
v = [1, 2]
w = [3, 4]
```

**Expected output:**
```python
[4, 6]
```

**Hint:** Use `zip()` and list comprehension.

<details>
<summary>Click to view solution</summary>

```python
def vector_add(v, w):
    """Adds corresponding elements of two vectors."""
    return [v_i + w_i for v_i, w_i in zip(v, w)]

# Test
v = [1, 2]
w = [3, 4]
print(vector_add(v, w))  # [4, 6]

# Verify with different lengths
print(vector_add([10, 20, 30], [5, 10, 15]))  # [15, 30, 45]
```

## Exercise 2 — Vector Subtraction

### Question

Write a function that subtracts one vector from another.

**Example:**
```python
v = [5, 6]
w = [2, 1]
```

**Expected output:**
```python
[3, 5]
```

**Hint:** Subtraction is addition of the negative: `v - w = v + (-w)`.

<details>
<summary>Click to view solution</summary>

```python
def vector_subtract(v, w):
    """Subtracts corresponding elements of two vectors."""
    return [v_i - w_i for v_i, w_i in zip(v, w)]

# Test
v = [5, 6]
w = [2, 1]
print(vector_subtract(v, w))  # [3, 5]

# Verify: subtraction and addition of negative should match
result1 = vector_subtract(v, w)
result2 = vector_add(v, [-x for x in w])
print(f"Subtraction: {result1}")
print(f"Add negative: {result2}")
print(f"Same? {result1 == result2}")  # True
```

## Exercise 3 — Scalar Multiplication

### Question

Multiply a vector by a scalar (single number).

**Example:**
```python
scalar = 3
vector = [1, 2, 3]
```

**Expected output:**
```python
[3, 6, 9]
```

<details>
<summary>Click to view solution</summary>

```python
def scalar_multiply(c, v):
    """Multiplies every element of a vector by a scalar."""
    return [c * v_i for v_i in v]

# Test
print(scalar_multiply(3, [1, 2, 3]))      # [3, 6, 9]
print(scalar_multiply(0.5, [1, 2, 3]))    # [0.5, 1.0, 1.5]
print(scalar_multiply(-1, [1, 2, 3]))     # [-1, -2, -3]
print(scalar_multiply(0, [1, 2, 3]))      # [0, 0, 0]
```

## Exercise 4 — Vector Sum and Mean

### Question

Implement `vector_sum` (adds all vectors together) and `vector_mean` (computes componentwise average).

**Example:**
```python
vectors = [[1, 2], [3, 4], [5, 6]]
```

**Expected output:**
```python
Sum:  [9, 12]
Mean: [3.0, 4.0]
```

**Hint:** Use `vector_add` for the sum, then `scalar_multiply(1/n, sum)` for the mean.

<details>
<summary>Click to view solution</summary>

```python
def vector_sum(vectors):
    """Sums all corresponding elements across multiple vectors."""
    result = vectors[0]
    for vector in vectors[1:]:
        result = vector_add(result, vector)
    return result

def vector_mean(vectors):
    """Computes the componentwise mean of a list of vectors."""
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum(vectors))

# Test
vectors = [[1, 2], [3, 4], [5, 6]]
print(f"Sum: {vector_sum(vectors)}")      # [9, 12]
print(f"Mean: {vector_mean(vectors)}")    # [3.0, 4.0]

# Verify mean calculation
print(f"Verify: (1+3+5)/3 = {(1+3+5)/3}, (2+4+6)/3 = {(2+4+6)/3}")
```

## Exercise 5 — Dot Product

### Question

Implement the dot product manually and explain what it represents.

**Example:**
```python
v = [1, 2]
w = [3, 4]
```

**Expected output:**
```python
11
```

**Interpretation question:** What does a large positive dot product mean? What about zero? Negative?

<details>
<summary>Click to view solution</summary>

```python
def dot(v, w):
    """Computes the dot product of two vectors.
    
    The dot product measures how much two vectors point in the same direction.
    - Large positive → similar direction
    - Zero → perpendicular
    - Negative → opposite directions
    """
    return sum(v_i * w_i for v_i, w_i in zip(v, w))

# Test
v = [1, 2]
w = [3, 4]
print(f"dot([1,2], [3,4]) = {dot(v, w)}")  # 11
print(f"  Calculation: (1×3) + (2×4) = 3 + 8 = 11")

# Explore similarity
print(f"\ndot([1,0], [2,0]) = {dot([1,0], [2,0])}")    # 2 (same direction)
print(f"dot([1,0], [0,1]) = {dot([1,0], [0,1])}")      # 0 (perpendicular)
print(f"dot([1,0], [-1,0]) = {dot([1,0], [-1,0])}")    # -1 (opposite)

# ML application: similarity between vectors
user_embedding = [0.8, 0.3, 0.1]  # User preferences
item_embedding = [0.7, 0.4, 0.2]  # Item features
similarity = dot(user_embedding, item_embedding)
print(f"\nUser-item similarity: {similarity:.3f}")
```

## Exercise 6 — Magnitude of a Vector

### Question

Find the magnitude (length) of a vector.

**Example:**
```python
[3, 4]
```

**Expected output:**
```python
5.0
```

**Hint:** Use the Pythagorean theorem: √(x² + y²) for 2D, √(x² + y² + z²) for 3D.

<details>
<summary>Click to view solution</summary>

```python
import math

def sum_of_squares(v):
    """Computes v₁² + v₂² + ... + vₙ²"""
    return dot(v, v)

def magnitude(v):
    """Computes the magnitude (length) of a vector."""
    return math.sqrt(sum_of_squares(v))

# Test
print(f"| [3, 4] | = {magnitude([3, 4])}")           # 5.0
print(f"| [1, 1, 1] | = {magnitude([1, 1, 1]):.3f}")  # 1.732

# Verify: magnitude of unit vectors
print(f"\n| [1, 0] | = {magnitude([1, 0])}")        # 1.0
print(f"| [0, 1] | = {magnitude([0, 1])}")          # 1.0

# Magnitude shows vector "size"
print(f"\nSmall vector: | [0.1, 0.2] | = {magnitude([0.1, 0.2]):.3f}")
print(f"Large vector: | [10, 20] | = {magnitude([10, 20]):.3f}")
```

## Exercise 7 — Euclidean Distance

### Question

Find the distance between two vectors.

**Example:**
```python
v = [1, 2]
w = [4, 6]
```

**Expected output:**
```python
5.0
```

**Interpretation question:** Why is distance important in machine learning?

<details>
<summary>Click to view solution</summary>

```python
import math

def squared_distance(v, w):
    """Computes the squared Euclidean distance between two vectors."""
    return sum_of_squares(vector_subtract(v, w))

def distance(v, w):
    """Computes the Euclidean distance between two vectors.
    
    This measures how far apart two points are in vector space.
    Used in KNN, clustering, recommendation systems, anomaly detection.
    """
    return math.sqrt(squared_distance(v, w))
    # Alternative: return magnitude(vector_subtract(v, w))

# Test
v = [1, 2]
w = [4, 6]
print(f"Distance: {distance(v, w)}")  # 5.0
print(f"  Calculation: √((4-1)² + (6-2)²) = √(9 + 16) = 5")

# Practical: find closest point
points = [[3, 4], [1, 1], [5, 0], [0, 2]]
origin = [0, 0]

print(f"\nDistance from origin:")
for point in points:
    d = distance(origin, point)
    print(f"  {point}: {d:.3f}")

# Find nearest
closest = min(points, key=lambda p: distance(origin, p))
print(f"\nClosest to origin: {closest}")

# Verify: distance = magnitude of difference
print(f"\nVerification:")
d1 = distance(v, w)
d2 = magnitude(vector_subtract(v, w))
print(f"  distance(v, w) = {d1}")
print(f"  magnitude(v - w) = {d2}")
print(f"  Equal? {d1 == d2}")
```

## Exercise 8 — Matrix Shape

### Question

Write a function that returns `(number of rows, number of columns)` for a matrix.

**Example:**
```python
matrix = [
    [1, 2, 3],
    [4, 5, 6]
]
```

**Expected output:**
```python
(2, 3)
```

<details>
<summary>Click to view solution</summary>

```python
def shape(matrix):
    """Returns the shape of a matrix as (num_rows, num_cols)."""
    num_rows = len(matrix)
    num_cols = len(matrix[0]) if matrix else 0
    return num_rows, num_cols

# Test
A = [[1, 2, 3], [4, 5, 6]]
B = [[1, 2], [3, 4], [5, 6]]
C = [[1, 2, 3, 4]]
D = []

print(f"A shape: {shape(A)}")  # (2, 3)
print(f"B shape: {shape(B)}")  # (3, 2)
print(f"C shape: {shape(C)}")  # (1, 4) - row vector
print(f"D shape: {shape(D)}")  # (0, 0) - empty matrix
```

## Exercise 9 — Get Matrix Row and Column

### Question

Write `get_row()` and `get_column()` functions to extract specific rows and columns.

**Example:**
```python
A = [
    [1, 2, 3],
    [4, 5, 6]
]
```

**Expected output:**
```python
get_row(A, 0)     → [1, 2, 3]
get_column(A, 1)  → [2, 5]
```

<details>
<summary>Click to view solution</summary>

```python
def get_row(matrix, i):
    """Returns the i-th row of a matrix."""
    return matrix[i]

def get_column(matrix, j):
    """Returns the j-th column of a matrix as a list."""
    return [row[j] for row in matrix]

# Test
A = [[1, 2, 3], [4, 5, 6]]

print(f"Matrix:")
for row in A:
    print(f"  {row}")

print(f"\nRow 0: {get_row(A, 0)}")       # [1, 2, 3]
print(f"Row 1: {get_row(A, 1)}")         # [4, 5, 6]
print(f"Column 0: {get_column(A, 0)}")   # [1, 4]
print(f"Column 1: {get_column(A, 1)}")   # [2, 5]
print(f"Column 2: {get_column(A, 2)}")   # [3, 6]

# Note: Row access is O(1), column access is O(num_rows)
```

## Exercise 10 — Dynamic Matrix Generation

### Question

Implement `make_matrix()` that creates a matrix using a function to define each entry. Use it to create a 5×5 identity matrix.

**Hint:** Nested list comprehension. For identity: `entry_fn(i, j)` returns `1` when `i == j`, else `0`.

<details>
<summary>Click to view solution</summary>

```python
def make_matrix(num_rows, num_cols, entry_fn):
    """
    Creates a matrix where entry_fn(i, j) defines the value at position (i, j).
    """
    return [[entry_fn(i, j) for j in range(num_cols)]
            for i in range(num_rows)]

# Identity matrix
identity = make_matrix(5, 5, lambda i, j: 1 if i == j else 0)

print("5×5 Identity Matrix:")
for row in identity:
    print(f"  {row}")

# All-ones matrix
print(f"\n3×4 All-Ones Matrix:")
ones = make_matrix(3, 4, lambda i, j: 1)
for row in ones:
    print(f"  {row}")

# Multiplication table
print(f"\n4×4 Multiplication Table:")
mult_table = make_matrix(4, 4, lambda i, j: (i+1) * (j+1))
for row in mult_table:
    print(f"  {row}")
```

Convert an undirected edge list into an adjacency matrix. Verify that you can check connections and extract neighbors.

**Data:**
```python
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
               (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

**Hint:** Initialize a zero matrix, then set `matrix[i][j] = matrix[j][i] = 1` for each edge.

<details>
<summary>Click to view solution</summary>

```python
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
               (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]

# Build 10×10 adjacency matrix
adj_matrix = make_matrix(10, 10, lambda i, j: 0)

for i, j in friendships:
    adj_matrix[i][j] = 1
    adj_matrix[j][i] = 1  # Undirected graph

# Verify O(1) connection lookups
print(f"Are users 0 and 2 connected? {adj_matrix[0][2] == 1}")   # True
print(f"Are users 0 and 8 connected? {adj_matrix[0][8] == 1}")   # False
print(f"Are users 5 and 7 connected? {adj_matrix[5][7] == 1}")   # True

# Extract neighbors of user 5
friends_of_5 = [i for i, is_friend in enumerate(adj_matrix[5]) if is_friend]
print(f"\nFriends of user 5: {friends_of_5}")  # [4, 6, 7]

# Find users with most friends
degree = [sum(row) for row in adj_matrix]
most_popular = degree.index(max(degree))
print(f"\nUser with most friends: User {most_popular} with {degree[most_popular]} friends")
```

## Exercise 12 — Real-World ML Thinking

### Question

A student is represented as:
```python
[20, 5, 85]  # [age, study_hours, marks]
```

Answer these questions:

1. Why is this a vector?
2. How would multiple students become a matrix?
3. Why might this representation be useful in ML?
4. How could the dot product be used with student vectors?

<details>
<summary>Click to view solution</summary>

**Discussion:**

**1. Why is this a vector?**
A student becomes a vector because their features (age, study hours, marks) are represented as an ordered collection of numbers. Each dimension represents a different attribute.

**2. Multiple students become a matrix:**
```python
students = [
    [20, 5, 85],   # Student 1: age=20, study_hours=5, marks=85
    [22, 6, 90],   # Student 2
    [19, 3, 70],   # Student 3
    [21, 8, 95]    # Student 4
]
# This is a 4×3 matrix (4 students, 3 features)
```

**3. Why useful in ML?**
ML models work with numerical feature representations:
- Each student is a data point in 3D feature space
- Similar students are close together in this space
- Models can learn patterns: "more study hours → higher marks"
- Distance metrics can find similar students
- The matrix format enables batch processing

**4. How to use dot product with student vectors:**
```python
# Compare two students
student_a = [20, 5, 85]
student_b = [22, 6, 90]

similarity = dot(student_a, student_b)
print(f"Raw dot product: {similarity}")

# Better: normalize and use cosine similarity
mag_a = magnitude(student_a)
mag_b = magnitude(student_b)
cosine_sim = dot(student_a, student_b) / (mag_a * mag_b)
print(f"Cosine similarity: {cosine_sim:.3f}")
```
