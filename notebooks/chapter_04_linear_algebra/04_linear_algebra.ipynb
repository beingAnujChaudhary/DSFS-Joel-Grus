## 1. What is a Vector?

A vector is:

> An ordered collection of numbers.

Example:
```python
[1, 2, 3]
```

Vectors can represent:
- Features of an observation (age, height, weight)
- Positions in space (x, y coordinates)
- Embeddings (vector representations of words, images, users)
- Directions and magnitudes


````python
# Example vectors

# Physical measurements: [height_cm, weight_kg, age_years]
height_weight_age = [170, 70, 21]

# Academic scores: [math, science, english]
student_scores = [85, 90, 78]

# 2D position: [x, y]
point_2d = [3, 4]

# 3D position: [x, y, z]
point_3d = [1, 2, 3]

print("Physical measurements:", height_weight_age)
print("Student scores:", student_scores)
print("2D point:", point_2d)
print("3D point:", point_3d)
````

**ML Connection:** A user profile might be `[age, income, engagement_score]` — this vector helps recommendation systems find similar users.

## 2. Vector Addition

Vector addition combines information from two vectors.

```python
[1, 2] + [3, 4]  →  [4, 6]
```


````python
def vector_add(v, w):
    """Adds corresponding elements of two vectors."""
    return [v_i + w_i for v_i, w_i in zip(v, w)]

# Test
v = [1, 2]
w = [3, 4]

print(f"{v} + {w} = {vector_add(v, w)}")

# Another example
a = [10, 20, 30]
b = [5, 10, 15]
print(f"{a} + {b} = {vector_add(a, b)}")

# Verify: vectors must be same length
try:
    vector_add([1, 2], [3, 4, 5])
except Exception as e:
    print(f"Error with different lengths: vectors must have same dimensions")
````

**ML Connection:** Combining feature vectors from different sources (e.g., adding a bias term to neural network inputs).

## 3. Vector Subtraction

Measures the difference between vectors — useful for finding how far apart two data points are.

```python
[5, 6] - [2, 1]  →  [3, 5]
```


````python
def vector_subtract(v, w):
    """Subtracts corresponding elements of two vectors."""
    return [v_i - w_i for v_i, w_i in zip(v, w)]

# Test
v = [5, 6]
w = [2, 1]

print(f"{v} - {w} = {vector_subtract(v, w)}")

# Another example: comparing two students
student_a = [85, 90, 78]  # math, science, english
student_b = [75, 82, 88]

difference = vector_subtract(student_a, student_b)
print(f"\nStudent A: {student_a}")
print(f"Student B: {student_b}")
print(f"Difference (A - B): {difference}")
print(f"  → Student A scored {difference[0]} more in math")
print(f"  → Student A scored {difference[1]} more in science")
print(f"  → Student A scored {difference[2]} less in english")
````

**ML Connection:** Computing gradients in optimization (difference between predicted and actual values).

## 4. Scalar Multiplication

A scalar is a single number. Scalar multiplication scales a vector — making it longer or shorter.

```python
2 * [1, 2, 3]  →  [2, 4, 6]
```


````python
def scalar_multiply(c, v):
    """Multiplies every element of a vector by a scalar."""
    return [c * v_i for v_i in v]

# Test
vector = [1, 2, 3]

print(f"Original vector: {vector}")
print(f"2 × vector: {scalar_multiply(2, vector)}")
print(f"0.5 × vector: {scalar_multiply(0.5, vector)}")
print(f"-1 × vector: {scalar_multiply(-1, vector)}")
print(f"0 × vector: {scalar_multiply(0, vector)}")
````

**ML Connection:**
- Scaling features during normalization
- Adjusting learning rates in gradient descent
- Weight updates in neural networks

## 5. Vector Sum and Mean

The sum combines multiple vectors. The mean computes the average vector — useful for finding the "center" of a dataset.



````python
def vector_sum(vectors):
    """Sums all corresponding elements across multiple vectors."""
    result = vectors[0]
    for vector in vectors[1:]:
        result = vector_add(result, vector)
    return result

def vector_mean(vectors):
    """Computes the componentwise mean of a list of vectors."""
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum(vectors))

# Test
vectors = [
    [1, 2],
    [3, 4],
    [5, 6]
]

print(f"Vectors: {vectors}")
print(f"Sum: {vector_sum(vectors)}")
print(f"Mean: {vector_mean(vectors)}")

# Verification
print(f"\nVerification:")
print(f"  Mean x-coordinate: (1 + 3 + 5) / 3 = {(1+3+5)/3}")
print(f"  Mean y-coordinate: (2 + 4 + 6) / 3 = {(2+4+6)/3}")
````

**ML Connection:** Computing centroids in k-means clustering; batch normalization in neural networks.

## 6. Dot Product

The dot product measures **how much two vectors point in the same direction**. It is one of the most important operations in machine learning.

```python
[1, 2] · [3, 4] = (1×3) + (2×4) = 3 + 8 = 11
```


````python
def dot(v, w):
    """Computes the dot product of two vectors."""
    return sum(v_i * w_i for v_i, w_i in zip(v, w))

# Test
v = [1, 2]
w = [3, 4]

print(f"v = {v}")
print(f"w = {w}")
print(f"dot(v, w) = {dot(v, w)}")
print(f"  Calculation: (1×3) + (2×4) = 3 + 8 = 11")

# Compare similar vs dissimilar vectors
similar_a = [1, 0]
similar_b = [2, 0]  # Points in same direction
print(f"\nSimilar vectors:")
print(f"  dot([1,0], [2,0]) = {dot([1,0], [2,0])} (large → similar direction)")

different_a = [1, 0]
different_b = [0, 1]  # Perpendicular
print(f"Perpendicular vectors:")
print(f"  dot([1,0], [0,1]) = {dot([1,0], [0,1])} (zero → perpendicular)")

opposite_a = [1, 0]
opposite_b = [-1, 0]  # Opposite direction
print(f"Opposite vectors:")
print(f"  dot([1,0], [-1,0]) = {dot([1,0], [-1,0])} (negative → opposite direction)")
````

### Why the Dot Product Matters

Used extensively in:
- **Recommendation systems**: similarity between user and item embeddings
- **Search ranking**: relevance between query and document vectors
- **Embeddings**: measuring semantic similarity between words
- **Neural networks**: weighted sums in each neuron
- **Transformers**: attention scores between query and key vectors (scaled dot-product attention)
- **Cosine similarity**: normalized dot product for comparing document similarity

## 7. Magnitude (Vector Length)

Magnitude represents the length of a vector — how far the point is from the origin.

For `[3, 4]`: √(3² + 4²) = √25 = 5



````python
import math

def sum_of_squares(v):
    """Computes the sum of squared elements: v₁² + v₂² + ... + vₙ²"""
    return dot(v, v)

def magnitude(v):
    """Computes the magnitude (length) of a vector."""
    return math.sqrt(sum_of_squares(v))

# Test
vector = [3, 4]
print(f"Vector: {vector}")
print(f"Sum of squares: {sum_of_squares(vector)}")
print(f"  Calculation: 3² + 4² = 9 + 16 = 25")
print(f"Magnitude: {magnitude(vector)}")
print(f"  Calculation: √25 = 5")

# Compare vectors of different lengths
vectors_to_compare = [
    [1, 0],
    [3, 4],
    [1, 1, 1],
    [10, 0, 0]
]

print(f"\nVector → Magnitude")
for v in vectors_to_compare:
    print(f"  {str(v):<15} → {magnitude(v):.3f}")
````


**ML Connection:**
- Normalizing feature vectors to unit length
- Computing regularization penalties (L2 norm)
- Measuring model parameter sizes

## 8. Distance

Distance measures how far apart two vectors are. Euclidean distance is the straight-line distance between two points.


````python
def squared_distance(v, w):
    """Computes the squared Euclidean distance between two vectors."""
    return sum_of_squares(vector_subtract(v, w))

def distance(v, w):
    """Computes the Euclidean distance between two vectors."""
    return math.sqrt(squared_distance(v, w))
    # Alternative: return magnitude(vector_subtract(v, w))

# Test
v = [1, 2]
w = [4, 6]

print(f"v = {v}")
print(f"w = {w}")
print(f"Distance between v and w: {distance(v, w):.3f}")
print(f"  Calculation: √((4-1)² + (6-2)²) = √(9 + 16) = √25 = 5")

# Practical example: find the closest point
reference = [0, 0]
points = [
    [3, 4],
    [1, 1],
    [5, 0],
    [0, 2]
]

print(f"\nDistance from origin [0, 0]:")
for point in points:
    d = distance(reference, point)
    print(f"  {point} → distance = {d:.3f}")

# Find closest
closest = min(points, key=lambda p: distance(reference, p))
print(f"  Closest point to origin: {closest}")
````

**ML Connection:**
- k-Nearest Neighbors (KNN) classification
- k-means clustering
- Recommendation systems (find similar users/items)
- Anomaly detection (points far from normal clusters)

## 9. What is a Matrix?

A matrix is:

> A collection of vectors arranged into rows and columns.

Think of it as a 2D table of numbers.



````python
matrix = [
    [1, 2, 3],   # Row 0
    [4, 5, 6]    # Row 1
]

print("Matrix:")
for row in matrix:
    print(f"  {row}")

print(f"\nThis is a 2 × 3 matrix (2 rows, 3 columns)")
````

### Matrix Dimensions

Matrix shape is expressed as `rows × columns`:


````python
def shape(matrix):
    """Returns the shape of a matrix as (num_rows, num_cols)."""
    num_rows = len(matrix)
    num_cols = len(matrix[0]) if matrix else 0
    return num_rows, num_cols

# Test
A = [
    [1, 2, 3],
    [4, 5, 6]
]

print(f"Matrix A:")
for row in A:
    print(f"  {row}")
print(f"Shape: {shape(A)} → {shape(A)[0]} rows × {shape(A)[1]} columns")

# Different sizes
B = [[1, 2], [3, 4], [5, 6]]
print(f"\nMatrix B (3×2):")
for row in B:
    print(f"  {row}")
print(f"Shape: {shape(B)}")

# Single row (vector as matrix)
C = [[1, 2, 3, 4]]
print(f"\nRow vector (1×4): {C}")
print(f"Shape: {shape(C)}")

# Single column
D = [[1], [2], [3]]
print(f"\nColumn vector (3×1): {D}")
print(f"Shape: {shape(D)}")
````

---

## 10. Row and Column Access


````python
def get_row(matrix, i):
    """Returns the i-th row of a matrix."""
    return matrix[i]

def get_column(matrix, j):
    """Returns the j-th column of a matrix as a list."""
    return [row[j] for row in matrix]

# Test
A = [
    [1, 2, 3],
    [4, 5, 6]
]

print("Matrix A:")
for row in A:
    print(f"  {row}")

print(f"\nRow 0: {get_row(A, 0)}")
print(f"Row 1: {get_row(A, 1)}")
print(f"Column 0: {get_column(A, 0)}")
print(f"Column 1: {get_column(A, 1)}")
print(f"Column 2: {get_column(A, 2)}")
````

## 11. Creating Matrices

````python
def make_matrix(num_rows, num_cols, entry_fn):
    """
    Creates a matrix of size num_rows × num_cols
    where entry_fn(i, j) defines the value at position (i, j).
    """
    return [[entry_fn(i, j) for j in range(num_cols)]
            for i in range(num_rows)]

# Identity matrix (1s on diagonal, 0s elsewhere)
def identity_matrix(n):
    return make_matrix(n, n, lambda i, j: 1 if i == j else 0)

print("3×3 Identity Matrix:")
I = identity_matrix(3)
for row in I:
    print(f"  {row}")

# All-ones matrix
print("\n2×4 All-Ones Matrix:")
ones = make_matrix(2, 4, lambda i, j: 1)
for row in ones:
    print(f"  {row}")

# Diagonal matrix
print("\n3×3 Diagonal Matrix [1, 2, 3]:")
diagonal = make_matrix(3, 3, lambda i, j: (i+1) if i == j else 0)
for row in diagonal:
    print(f"  {row}")
````
