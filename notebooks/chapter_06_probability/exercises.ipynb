## Exercise 1 — Coin Flip Probability

### Question

What is the probability of getting Heads from a fair coin?

**Interpretation question:** Why is the probability 0.5?

<details>
<summary>Click to view solution</summary>

For a fair coin:

**Sample space:** {Heads, Tails}

**Probability:** 1 / 2 = 0.5

Because both outcomes are equally likely in a fair coin. The sample space has exactly 2 equally probable outcomes, and "Heads" is exactly 1 of them.

```python
# Verification
sample_space = ["Heads", "Tails"]
favorable = ["Heads"]
probability = len(favorable) / len(sample_space)
print(f"P(Heads) = {len(favorable)}/{len(sample_space)} = {probability}")
```
</details>

## Exercise 2 — Simulate Coin Flips

### Question

Simulate 100 coin flips and count Heads and Tails.

**Interpretation question:** Why do results approach 50/50 over time?

<details>
<summary>Click to view solution</summary>

```python
import random

heads = 0
tails = 0

for _ in range(100):
    flip = random.choice(["Heads", "Tails"])
    if flip == "Heads":
        heads += 1
    else:
        tails += 1

print(f"Heads: {heads} ({heads}%)")
print(f"Tails: {tails} ({tails}%)")
```

**Why results approach 50/50:** Because of the **Law of Large Numbers** — as the number of trials increases, the observed frequency approaches the true probability. With only 10 flips, you might see 7 Heads (70%), but with 10,000 flips, the proportion will be much closer to 50%.

## Exercise 4 — Conditional Probability Intuition

### Question

Suppose:
- Probability of rain = 30%
- But dark clouds appear
- Now probability of rain = 70%

**Interpretation question:** Why did the probability change?

<details>
<summary>Click to view solution</summary>

The probability changed because **new evidence appeared** (dark clouds).

Conditional probability updates likelihood using additional information:
- **Prior:** P(Rain) = 0.30 (before seeing clouds)
- **Posterior:** P(Rain | Clouds) = 0.70 (after seeing clouds)

The condition (clouds) provides relevant information that changes our assessment. This is the foundation of Bayesian thinking — update beliefs when new evidence arrives.

```python
# Conceptual example
prior_rain = 0.30
posterior_rain_given_clouds = 0.70

print(f"Before evidence: P(Rain) = {prior_rain:.0%}")
print(f"After evidence:  P(Rain|Clouds) = {posterior_rain_given_clouds:.0%}")
print(f"  → New evidence updated our belief!")
```
</details>

## Exercise 5 — Independent Events

### Question

Are these events independent?

**Case 1:** Coin flip 1 and Coin flip 2

**Case 2:** Drawing two cards from a deck without replacement

**Explain why** for each case.

<details>
<summary>Click to view solution</summary>

**Case 1: Independent.** One coin flip does not affect another. The outcome of flip 1 gives no information about flip 2.
- P(Heads on flip 2 | Heads on flip 1) = P(Heads on flip 2) = 0.5

**Case 2: Dependent.** Drawing without replacement means earlier draws change the remaining possibilities.
- P(Ace on 2nd draw | Ace on 1st draw) = 3/51
- P(Ace on 2nd draw without condition) = 4/52
- These are different → dependent events

```python
# Demonstration: Case 1 (Independent)
flip_1 = random.choice(["Heads", "Tails"])
flip_2 = random.choice(["Heads", "Tails"])
print(f"Coin flips: {flip_1}, {flip_2}")
print("  → Independent: first flip doesn't affect the second")

# Demonstration: Case 2 (Dependent)
deck = ['A♠', 'K♠', 'Q♠', 'J♠', '10♠']  # Simplified deck
first_draw = deck.pop(random.randint(0, len(deck)-1))
second_draw = random.choice(deck)
print(f"\nCard draws: {first_draw}, {second_draw}")
print(f"  → Dependent: first draw changed available cards")
print(f"  → Remaining deck: {deck}")
```

## Exercise 6 — Simulating Conditional Probability (Two Children Paradox)

### Question

Simulate the "Two Children" paradox using 10,000 Monte Carlo simulations.

Calculate:
1. P(both girls | older child is a girl)
2. P(both girls | at least one child is a girl)

**Hint:** Use `random.choice(["boy", "girl"])` and track counters for each condition.

<details>
<summary>Click to view solution</summary>

```python
import random

def random_kid():
    return random.choice(["boy", "girl"])

both_girls = 0
older_girl = 0
either_girl = 0

random.seed(42)
for _ in range(10000):
    younger = random_kid()
    older = random_kid()
    
    if older == "girl":
        older_girl += 1
    if older == "girl" and younger == "girl":
        both_girls += 1
    if older == "girl" or younger == "girl":
        either_girl += 1

print(f"P(both girls | older is girl):     {both_girls / older_girl:.3f}  (Expected: ~0.500)")
print(f"P(both girls | at least one girl):  {both_girls / either_girl:.3f}  (Expected: ~0.333)")
print(f"\n🔍 'At least one' is a different condition than 'the older one'")
print(f"   → The sample space shrinks differently in each case")
```
</details>

## Exercise 7 — Bayes's Theorem for Medical Testing

### Question

Implement `bayes_theorem()` and solve: A disease affects 1 in 1,000 people. A test is 95% accurate (5% false positive rate). If someone tests positive, what's the probability they actually have the disease?

<details>
<summary>Click to view solution</summary>

```python
def bayes_theorem(p_e, p_f_given_e, p_f_given_not_e):
    """Calculates P(E|F) using Bayes's Theorem.
    
    P(E|F) = P(F|E) * P(E) / P(F)
    where P(F) = P(F|E)*P(E) + P(F|¬E)*P(¬E)
    """
    p_not_e = 1 - p_e
    p_f = p_f_given_e * p_e + p_f_given_not_e * p_not_e
    return (p_f_given_e * p_e) / p_f

# Medical Test Parameters
p_disease = 0.001                 # 1 in 1,000
p_positive_given_disease = 0.95   # 95% true positive rate
p_positive_given_healthy = 0.05   # 5% false positive rate

p_disease_given_positive = bayes_theorem(
    p_disease,
    p_positive_given_disease,
    p_positive_given_healthy
)

print(f"Given:")
print(f"  Disease prevalence: {p_disease:.3f} ({p_disease*100:.1f}%)")
print(f"  True positive rate: {p_positive_given_disease:.0%}")
print(f"  False positive rate: {p_positive_given_healthy:.0%}")
print(f"\nResult:")
print(f"  P(Disease | Positive Test) = {p_disease_given_positive:.4f}")
print(f"  That's only {p_disease_given_positive*100:.2f}%!")
print(f"\n🔍 Even with a 95% accurate test, a positive result")
print(f"   means <2% chance of actually having the disease!")
print(f"   → Base rates (prior probability) matter enormously")
```

## Exercise 8 — Expected Value

### Question

Calculate the expected value of a fair die.

**Hint:** Average all possible outcomes equally.

**Interpretation question:** Why is the expected value 3.5 even though it never appears on a die?

<details>
<summary>Click to view solution</summary>

```python
dice_values = [1, 2, 3, 4, 5, 6]
expected_value = sum(dice_values) / len(dice_values)

print(f"Expected value = ({'+'.join(map(str, dice_values))}) / {len(dice_values)}")
print(f"               = {sum(dice_values)} / {len(dice_values)}")
print(f"               = {expected_value}")
print(f"\nEven though 3.5 never appears on a single roll,")
print(f"it's the long-run average over many rolls.")
print(f"\nIf you roll a die 1,000,000 times, the average")
print(f"will converge to approximately {expected_value}")
```

**Why 3.5?** Expected value is a **long-run average**, not an observable outcome. It's the center of the probability distribution — like the balance point of a seesaw. It's useful for decision making even though any single observation will be an integer.

## Exercise 9 — Probability Range

### Question

Can probability ever be -1, 2, or 100? Why or why not?

<details>
<summary>Click to view solution</summary>

**No.** Probability always lies between **0 and 1** (inclusive).

- **0** → Impossible event (will never occur)
- **1** → Certain event (will always occur)
- Everything else lies between them

Probabilities cannot be negative because you can't have less than zero chance of something happening. They cannot exceed 1 because you can't be more than 100% certain.

If someone says "There's a 200% chance," they're either joking or using the term informally to mean "very likely" — not mathematically correct probability.

```python
# Valid probabilities
valid = [0, 0.5, 1]

# Invalid (not probabilities in the mathematical sense)
invalid = [-1, 2, 100]

print("Valid probabilities (0 to 1):")
for p in valid:
    print(f"  P(event) = {p} ✓")

print("\nInvalid 'probabilities':")
for p in invalid:
    print(f"  'P(event)' = {p} ✗ (outside [0,1] range)")
```
</details>

## Exercise 10 — Correlation ≠ Causation

### Question

Suppose ice cream sales increase and drowning incidents also increase. Does ice cream cause drowning? Why or why not?

<details>
<summary>Click to view solution</summary>

**No, ice cream does not cause drowning.**

This is a classic example of **spurious correlation** — two variables appear related but aren't causally connected.

**The hidden variable:** Summer weather (temperature)

- Hot weather → More ice cream sales
- Hot weather → More people swimming → More drowning incidents

Both increase because of a **confounding variable** (temperature), not because one causes the other.

**Key lesson:**
> Correlation does not imply causation.

Just because two variables move together doesn't mean one causes the other. Always consider confounding variables and use controlled experiments (A/B tests, randomized trials) to establish causation.

```python
# This pattern appears correlated but isn't causal
print("Observed pattern:")
print("  Ice cream sales ↑  ←→  Drowning incidents ↑")
print("\nActual explanation:")
print("  Hot weather → Ice cream sales ↑")
print("  Hot weather → More swimming → Drowning ↑")
print("\n  → Both caused by temperature (confounding variable)")
print("  → No direct causal link between ice cream and drowning")
```
</details>

## Exercise 11 — Continuous Distributions (Normal PDF & CDF)

### Question

Implement `normal_pdf()` and `normal_cdf()` from scratch. Plot both on side-by-side subplots comparing standard Normal (μ=0, σ=1) with shifted Normal (μ=2, σ=2).

**Hint:** Use `math.exp`, `math.sqrt`, and `math.erf`.

<details>
<summary>Click to view solution</summary>

```python
import math
import matplotlib.pyplot as plt

def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x - mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))

def normal_cdf(x, mu=0, sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2

# Plot
xs = [x / 10.0 for x in range(-50, 80)]

plt.figure(figsize=(14, 5))

# PDFs
plt.subplot(1, 2, 1)
plt.plot(xs, [normal_pdf(x, mu=0, sigma=1) for x in xs], '-', label='μ=0, σ=1 (standard)')
plt.plot(xs, [normal_pdf(x, mu=2, sigma=2) for x in xs], '--', label='μ=2, σ=2 (shifted, wider)')
plt.legend()
plt.title("Normal PDFs", fontweight='bold')
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)

# CDFs
plt.subplot(1, 2, 2)
plt.plot(xs, [normal_cdf(x, mu=0, sigma=1) for x in xs], '-', label='μ=0, σ=1')
plt.plot(xs, [normal_cdf(x, mu=2, sigma=2) for x in xs], '--', label='μ=2, σ=2')
plt.legend(loc='lower right')
plt.title("Normal CDFs", fontweight='bold')
plt.xlabel("x")
plt.ylabel("P(X ≤ x)")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("PDF: Shows density — height of the bell curve at each point")
print("CDF: Shows accumulated probability — P(X ≤ x)")
```

## Exercise 12 — Inverting the Normal CDF (Binary Search)

### Question

The Normal CDF has no closed-form inverse. Implement `inverse_normal_cdf(p)` using binary search to find z-scores for the 90th, 95th, and 99th percentiles.

**Hint:** Start with bounds `low_z = -10.0` and `hi_z = 10.0`. Halve the search range until tolerance is reached.

<details>
<summary>Click to view solution</summary>

```python
def inverse_normal_cdf(p, mu=0, sigma=1, tolerance=0.00001):
    """Find approximate inverse using binary search."""
    if mu != 0 or sigma != 1:
        return mu + sigma * inverse_normal_cdf(p, tolerance=tolerance)
    
    low_z, low_p = -10.0, 0     # normal_cdf(-10) ≈ 0
    hi_z, hi_p   =  10.0, 1     # normal_cdf(10)  ≈ 1
    
    while hi_z - low_z > tolerance:
        mid_z = (low_z + hi_z) / 2
        mid_p = normal_cdf(mid_z)
        
        if mid_p < p:
            low_z, low_p = mid_z, mid_p
        elif mid_p > p:
            hi_z, hi_p = mid_z, mid_p
        else:
            break
    
    return mid_z

# Test
percentiles = [0.50, 0.90, 0.95, 0.99, 0.999]
print("Z-scores for common percentiles:")
for p in percentiles:
    z = inverse_normal_cdf(p)
    print(f"  {int(p*100)}th percentile: z = {z:.3f}")

print(f"\n🔍 When math can't provide a clean formula,")
print(f"   computer science (binary search) solves it!")
```
</details>

## Exercise 13 — The Central Limit Theorem in Action

### Question

Demonstrate the CLT: Simulate 10,000 trials of flipping a biased coin (p=0.6) 100 times. Plot the histogram of results with the Normal approximation overlay.

<details>
<summary>Click to view solution</summary>

```python
from collections import Counter

def bernoulli_trial(p):
    return 1 if random.random() < p else 0

def binomial(n, p):
    return sum(bernoulli_trial(p) for _ in range(n))

def make_hist(p, n, num_points):
    data = [binomial(n, p) for _ in range(num_points)]
    
    # Histogram of actual binomial samples
    histogram = Counter(data)
    plt.bar([x - 0.4 for x in histogram.keys()],
            [v / num_points for v in histogram.values()],
            0.8, color='steelblue', alpha=0.7, edgecolor='black',
            label='Binomial Samples')
    
    # Normal approximation
    mu = p * n
    sigma = math.sqrt(n * p * (1 - p))
    
    xs = range(min(data), max(data) + 1)
    ys = [normal_cdf(i + 0.5, mu, sigma) - normal_cdf(i - 0.5, mu, sigma)
          for i in xs]
    plt.plot(xs, ys, 'r-', linewidth=2, label='Normal Approximation')
    
    plt.title(f"Binomial(n={n}, p={p}) vs Normal(μ={mu}, σ={sigma:.1f})",
              fontsize=14, fontweight='bold')
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Visualize CLT
make_hist(0.6, 100, 10000)
print("The binomial distribution is closely approximated by the Normal!")
print(f"μ = np = {0.6*100}, σ = √(np(1-p)) = {math.sqrt(100*0.6*0.4):.1f}")
```
</details>

## Exercise 14 — Real-World ML Thinking

### Question

Why is probability important in machine learning? Think about spam filtering, recommendations, confidence scores, and uncertainty estimation.

<details>
<summary>Click to view solution</summary>

Probability helps ML systems:

**Classification:**
- "This email is 92% likely to be spam" — more informative than just "spam"
- Models output probabilities, not just binary decisions

**Recommendation Systems:**
- "User has 15% probability of clicking this item" — enables ranking
- Probabilities help balance exploration vs exploitation

**Confidence Estimation:**
- "The model is 95% confident in this prediction" — flags uncertain cases for human review
- Critical for high-stakes decisions (medical diagnosis, fraud detection)

**Uncertainty Modeling:**
- Modern AI systems estimate not just predictions but uncertainty bounds
- Bayesian neural networks, Monte Carlo dropout, calibration

**A/B Testing:**
- "There's a 99% probability that variant B is better than variant A"
- Statistical significance relies entirely on probability theory

**Without probability, ML would just be guessing — probability gives us a mathematical framework to quantify and reason about uncertainty.**

```python
# Conceptual example
print("Probability in ML systems:")
print("  Spam filter:    P(spam | email_content) = 0.92")
print("  Recommendation: P(user_clicks | item) = 0.15")
```
